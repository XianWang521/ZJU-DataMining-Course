## Let's try Spectral Clustering on real-world data!

```python
TDT2_data = sio.loadmat('TDT2_data.mat')
fea = TDT2_data['fea']
gnd = TDT2_data['gnd'].reshape(-1)
```

```python
from constructW import constructW
from bestMap import bestMap
from MutualInfo import MutualInfo
# YOUR CODE HERE
# begin answer
n = 100
SC_accuracy = 0
SC_mutInfo = 0
KM_accuracy = 0
KM_mutInfo = 0

for i in range(n):
    W = constructW(fea, k=10).toarray()
    SC_idx = bestMap(gnd, spectral(W, np.unique(gnd).shape[0]))
    KM_idx = bestMap(gnd, kmeans(x = fea.toarray(), k = np.unique(gnd).shape[0]))
    SC_accuracy += np.sum(gnd == SC_idx) / fea.shape[0]
    KM_accuracy += np.sum(gnd == KM_idx) / fea.shape[0]
    SC_mutInfo += MutualInfo(gnd, SC_idx)
    KM_mutInfo += MutualInfo(gnd, KM_idx)

SC_accuracy /= n
SC_mutInfo /= n
KM_accuracy /= n
KM_mutInfo /= n
print('spectral clustering accuracy:', SC_accuracy, 'normalized mutual information:', SC_mutInfo)
print('K-means clustering accuracy:', KM_accuracy, 'normalized mutual information:', KM_mutInfo)
# end answer

```
