```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

%load_ext autoreload
%autoreload 2
```

```python
digit_train = sio.loadmat('digit_train')
X = digit_train['X']
y = digit_train['y']
#print(X)
#print(y)
digit_test = sio.loadmat('digit_test')
X_test = digit_test['X_test']
y_test = digit_test['y_test']
```

```python
def show_digit(fea):
    plt.rcParams['figure.figsize'] = (10.0, 8.0)
    idx = np.random.permutation(X.shape[1])
    fea = fea[:, idx[:100]]
    faceW = 28
    faceH = 28
    numPerLine = 20
    ShowLine = 4
    Y = np.zeros((faceH * ShowLine, faceW * numPerLine), dtype=np.float)
    for i in range(ShowLine):
        for j in range(numPerLine):
            Y[i * faceH:(i + 1) * faceH, j * faceW:(j + 1) * faceW] = fea[:,i * numPerLine + j].reshape((faceH, faceW))            
    plt.imshow(Y, cmap='gray')
```

### Ridge Regression and LOOCV

```python
show_digit(X)
# Do feature normalization here
# here refer from https://medium.com/greyatom/why-how-and-when-to-scale-your-features-4b30ab09db5e
# begin answer
train_x = (X-(np.max(X)+np.min(X))/2)/(np.max(X)-np.min(X))
# end answer
# Do feature normalization to X_test, y_test
test_x = (X_test-(np.max(X_test)+np.min(X_test))/2)/(np.max(X_test)-np.min(X_test))

# Do LOOCV
lmbdas = np.array([1e-3, 1e-2, 1e-1, 0, 1, 1e1, 1e2, 1e3])
lmbda = 0
E_val_min = float('inf')

from ridge import ridge

for i in range(len(lmbdas)):
    E_val = 0
    for j in range(train_x.shape[1]):
        X_ = np.delete(train_x, j, axis=1)  # take point j out of X
        y_ = np.delete(y, j, axis=1)
        w = ridge(X_, y_, lmbdas[i])
        E_val += (y[0][j]-np.matmul(w.T,np.vstack((np.ones((1,1)), train_x[:,j].reshape(train_x.shape[0],1))))) ** 2
        
    # Update lmbda according validation error
    # begin answer
    if(E_val < E_val_min):
        lmbda = lmbdas[i]
        E_val_min = E_val
    # end answer
w = ridge(train_x, y, lmbda)
# Compute training error
w_y_train = np.sign(np.matmul(w.T, np.vstack((np.ones((1, train_x.shape[1])), train_x))))
E_train = np.sum(w_y_train != y)/train_x.shape[1]

# Compute test error
w_y_test = np.sign(np.matmul(w.T, np.vstack((np.ones((1, X_test.shape[1])), X_test))))
E_test = np.sum(w_y_test != y_test)/X_test.shape[1]

print(lmbda)
print('with', np.matmul(w.T, w)[0][0])
print('Train Error:', E_train, 'Test Error:', E_test)

w = ridge(train_x, y, 0)
# Compute training error
w_y_train = np.sign(np.matmul(w.T, np.vstack((np.ones((1, train_x.shape[1])), train_x))))
E_train = np.sum(w_y_train != y)/train_x.shape[1]

# Compute test error
w_y_test = np.sign(np.matmul(w.T, np.vstack((np.ones((1, X_test.shape[1])), X_test))))
E_test = np.sum(w_y_test != y_test)/X_test.shape[1]

print('without', np.matmul(w.T, w)[0][0])
print('Train Error:', E_train, 'Test Error:', E_test)
```

### Logistic Regression with Regularization
Use the simlimar skeleton code above to implement it.

```python
from logistic_r import logistic_r

for i in range(y.shape[1]):
    if(y[0][i] == -1):
        y[0][i] = 0
for i in range(y_test.shape[1]):
    if(y_test[0][i] == -1):
        y_test[0][i] = 0

# Do LOOCV
lmbdas = np.array([1e-3, 1e-2, 1e-1, 0, 1, 1e1, 1e2, 1e3])
lmbda = 0
E_val_min = float('inf')

for i in range(len(lmbdas)):
    E_val = 0
    for j in range(train_x.shape[1]):
        X_ = np.delete(train_x, j, axis=1)  # take point j out of X
        y_ = np.delete(y, j, axis=1)
        w = logistic_r(X_, y_, lmbdas[i])
        temp = np.matmul(w.T, np.vstack((np.ones((1,1)),train_x[:,j].reshape(train_x.shape[0],1))))
        if(temp[0][0] >= 0):
            w_x = 1.0/(1+np.exp(-temp))
        else:
            w_x = np.exp(temp)/(1+np.exp(temp))
        
        if(y[0][j] == 0):
            E_val -= np.log(1-w_x+1e-6)
        else:
            E_val -= np.log(w_x+1e-6)
    # Update lmbda according validation error
    # begin answer
    if(E_val < E_val_min):
        lmbda = lmbdas[i]
        E_val_min = E_val
    # end answer
    

w = logistic_r(train_x, y, lmbda)
# Compute training error
w_y_train = 1/(1+np.exp(-np.matmul(w.T, np.vstack((np.ones((1, train_x.shape[1])), train_x)))))

for i in range(w_y_train.shape[1]):
    if(w_y_train[0][i] > 0.5):
        w_y_train[0][i] = 1
    else:
        w_y_train[0][i] = 0
E_train = np.sum(w_y_train != y)/train_x.shape[1]

# Compute test error
w_y_test = 1/(1+np.exp(-np.matmul(w.T, np.vstack((np.ones((1, X_test.shape[1])), X_test)))))
for i in range(w_y_test.shape[1]):
    if(w_y_test[0][i] > 0.5):
        w_y_test[0][i] = 1
    else:
        w_y_test[0][i] = 0
E_test = np.sum(w_y_test != y_test)/X_test.shape[1]

print(lmbda)
print('with', np.matmul(w.T, w)[0][0])
print('Train Error:', E_train, 'Test Error:', E_test)

w = logistic_r(train_x, y, 0)
# Compute training error
w_y_train = 1/(1+np.exp(-np.matmul(w.T, np.vstack((np.ones((1, train_x.shape[1])), train_x)))))
for i in range(w_y_train.shape[1]):
    if(w_y_train[0][i] > 0.5):
        w_y_train[0][i] = 1
    else:
        w_y_train[0][i] = 0
E_train = np.sum(w_y_train != y)/train_x.shape[1]

# Compute test error
w_y_test = 1/(1+np.exp(-np.matmul(w.T, np.vstack((np.ones((1, X_test.shape[1])), X_test)))))
for i in range(w_y_test.shape[1]):
    if(w_y_test[0][i] > 0.5):
        w_y_test[0][i] = 1
    else:
        w_y_test[0][i] = 0
E_test = np.sum(w_y_test != y_test)/X_test.shape[1]

print('without', np.matmul(w.T, w)[0][0])
print('Train Error:', E_train, 'Test Error:', E_test)
```
